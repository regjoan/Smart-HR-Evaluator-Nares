### Feature Engineering Strategy and Limitations

Given the placeholder nature of the data, it's critical to acknowledge the following limitations and adjust our feature engineering strategy accordingly:

1.  **Identical DataFrames**: As observed during data loading and initial exploration, `attendance` and `communication` are currently identical, both loaded from `HR.csv`. Similarly, `tasks` and `feedback` are identical, both loaded from `WA_Fn-UseC_-HR-Employee-Attrition.csv`. 2.  **Merge Infeasibility**: A direct merge of all four DataFrames on a common `employee_id` is not conceptually sound or feasible. The `attendance`/`communication` DataFrames lack a clear employee identifier that can be mapped to the `employee_id` present in `tasks`/`feedback`. These datasets represent different employee populations or different aspects of employee data that cannot be directly joined in a meaningful way with the current placeholder setup.

3.  **Independent Processing**: Due to the merge limitations, feature engineering will proceed independently for two distinct groups of DataFrames:
    *   Group 1: `attendance` and `communication` (derived from `HR.csv`)
    *   Group 2: `tasks` and `feedback` (derived from `WA_Fn-UseC_-HR-Employee-Attrition.csv`)

4.  **Weak Proxies**: The engineered features will often serve as weaker proxies for the desired HR metrics. This is a direct consequence of using placeholder data that does not align with the original intended schema. This will be explicitly highlighted for each score.

Here's the strategy for calculating each score:

## Calculate Productivity Score

For `attendance` and `communication` DataFrames, calculate a 'Productivity_Score' using a combination of `number_project`, `average_montly_hours`, and `last_evaluation`. For `tasks` and `feedback` DataFrames, calculate 'Productivity_Score' using `PerformanceRating` and potentially `TotalWorkingYears` as proxies.


````python
from sklearn.preprocessing import MinMaxScaler

def min_max_scale(series):
    """Scales a pandas Series to a 0-1 range using MinMaxScaler."""
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(series.values.reshape(-1, 1))
    return pd.Series(scaled_data.flatten(), index=series.index)

print("Calculating Productivity_Score for attendance and communication DataFrames...")
# Process attendance and communication DataFrames
for df_name, df in [('attendance', attendance), ('communication', communication)]:
    print(f"\n--- Processing {df_name} DataFrame ---")
    required_cols_hr = ['number_project', 'average_montly_hours', 'last_evaluation']
    if all(col in df.columns for col in required_cols_hr):
        df['N_number_project'] = min_max_scale(df['number_project'])
        df['N_average_montly_hours'] = min_max_scale(df['average_montly_hours'])
        df['N_last_evaluation'] = min_max_scale(df['last_evaluation'])

        df['Productivity_Score'] = (0.35 * df['N_number_project'] +
                                    0.3 * df['N_average_montly_hours'] +
                                    0.35 * df['N_last_evaluation'])
        df.drop(columns=['N_number_project', 'N_average_montly_hours', 'N_last_evaluation'], inplace=True)
        print(f"'Productivity_Score' calculated and temporary columns dropped for {df_name}.")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Required columns {required_cols_hr} not found in {df_name}. Skipping Productivity_Score calculation.")

print("\nCalculating Productivity_Score for tasks and feedback DataFrames...")
# Process tasks and feedback DataFrames
for df_name, df in [('tasks', tasks), ('feedback', feedback)]:
    print(f"\n--- Processing {df_name} DataFrame ---")
    required_cols_attrition = ['PerformanceRating', 'TotalWorkingYears']
    if all(col in df.columns for col in required_cols_attrition):
        # Scale PerformanceRating manually as specified
        min_pr = df['PerformanceRating'].min()
        max_pr = df['PerformanceRating'].max()
        if (max_pr - min_pr) > 0: # Avoid division by zero if all values are the same
            df['N_PerformanceRating'] = (df['PerformanceRating'] - min_pr) / (max_pr - min_pr)
        else:
            df['N_PerformanceRating'] = 0.5 # Assign a neutral value if no variance

        df['N_TotalWorkingYears'] = min_max_scale(df['TotalWorkingYears'])

        df['Productivity_Score'] = (0.7 * df['N_PerformanceRating'] +
                                    0.3 * df['N_TotalWorkingYears'])
        df.drop(columns=['N_PerformanceRating', 'N_TotalWorkingYears'], inplace=True)
        print(f"'Productivity_Score' calculated and temporary columns dropped for {df_name}.")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Required columns {required_cols_attrition} not found in {df_name}. Skipping Productivity_Score calculation.")

print("Finished calculating Productivity_Score for all DataFrames.")
````

Following the feature engineering plan, I will now calculate the 'Punctuality_Score' for the `attendance` and `communication` DataFrames using `average_montly_hours` as an inverse proxy.
````python
print("\n--- Calculating Punctuality Score (with limitations) ---")

# For attendance and communication DataFrames: using average_montly_hours inversely
print("Calculating Punctuality_Score for attendance and communication (inverse of average_montly_hours)...")
for df_name, df in [('attendance', attendance), ('communication', communication)]:
    if 'average_montly_hours' in df.columns:
        # Less average_montly_hours is assumed to imply better punctuality/adherence to standard hours.
        # So, we'll inverse scale it. Max hours -> 0 punctuality, Min hours -> 1 punctuality.
        df['Punctuality_Score'] = 1 - min_max_scale(df['average_montly_hours'])
        print(f"'Punctuality_Score' calculated for {df_name} (inverse proxy).")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Skipping Punctuality_Score for {df_name}: 'average_montly_hours' not found.")

# For tasks and feedback DataFrames: No direct proxy available, explicitly state this limitation.
print("\nPunctuality_Score cannot be meaningfully created for tasks and feedback DataFrames due to lack of relevant columns.")
````

**Reasoning**:
Following the feature engineering plan, I will now calculate the 'Communication_Score' for the `attendance` and `communication` DataFrames using 'sales' as a weak proxy, and for the `tasks` and `feedback` DataFrames using 'JobInvolvement' and 'JobRole' as weak proxies. This involves assigning scores based on specific conditions and combining normalized values where applicable, as outlined in the subtask instructions.



````python
print("\n--- Calculating Communication Score (with limitations) ---")

# For attendance and communication DataFrames: using 'sales' as a very weak proxy
print("Calculating Communication_Score for attendance and communication (using 'sales' as proxy)...")
for df_name, df in [('attendance', attendance), ('communication', communication)]:
    if 'sales' in df.columns:
        # Assign higher communication score to 'sales' department
        df['Communication_Score'] = df['sales'].apply(lambda x: 1.0 if x == 'sales' else 0.5)
        print(f"'Communication_Score' calculated for {df_name} (weak proxy).")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Skipping Communication_Score for {df_name}: 'sales' column not found.")

# For tasks and feedback DataFrames: using JobRole categories and JobInvolvement
print("\nCalculating Communication_Score for tasks and feedback (using JobRole and JobInvolvement)...")
for df_name, df in [('tasks', tasks), ('feedback', feedback)]:
    if all(col in df.columns for col in ['JobInvolvement', 'JobRole']):
        # Assign scores based on JobRole, and combine with normalized JobInvolvement
        # Assuming 'Sales Representative' and 'Manager' roles require high communication.
        # Normalize JobInvolvement to be between 0 and 1
        df['N_JobInvolvement'] = min_max_scale(df['JobInvolvement'])
        df['JobRole_Comm_Factor'] = df['JobRole'].apply(lambda x: 1.0 if x in ['Sales Representative', 'Manager'] else (0.75 if x in ['Human Resources', 'Research Director', 'Healthcare Representative'] else 0.5))
        df['Communication_Score'] = (df['N_JobInvolvement'] * 0.5 + df['JobRole_Comm_Factor'] * 0.5)
        df.drop(columns=['N_JobInvolvement', 'JobRole_Comm_Factor'], inplace=True)
        print(f"'Communication_Score' calculated for {df_name} (weak proxy).")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Skipping Communication_Score for {df_name}: required columns not found.")
````

**Reasoning**:
Following the feature engineering plan, I will now calculate the 'Collaboration_Score' for the `attendance` and `communication` DataFrames using `number_project` as a weak proxy. For the `tasks` and `feedback` DataFrames, I will combine `JobInvolvement` and `RelationshipSatisfaction` to create the 'Collaboration_Score'.



````python
print("\n--- Calculating Collaboration Score (with limitations) ---")

# For attendance and communication DataFrames: using number_project as a weak proxy
print("Calculating Collaboration_Score for attendance and communication (using number_project)...")
for df_name, df in [('attendance', attendance), ('communication', communication)]:
    if 'number_project' in df.columns:
        # More projects imply more collaboration opportunities.
        df['Collaboration_Score'] = min_max_scale(df['number_project'])
        print(f"'Collaboration_Score' calculated for {df_name} (weak proxy).")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Skipping Collaboration_Score for {df_name}: 'number_project' not found.")

# For tasks and feedback DataFrames: using JobInvolvement and RelationshipSatisfaction
print("\nCalculating Collaboration_Score for tasks and feedback (using JobInvolvement and RelationshipSatisfaction)...")
for df_name, df in [('tasks', tasks), ('feedback', feedback)]:
    if all(col in df.columns for col in ['JobInvolvement', 'RelationshipSatisfaction']):
        # Combine normalized JobInvolvement and RelationshipSatisfaction
        df['N_JobInvolvement'] = min_max_scale(df['JobInvolvement'])
        df['N_RelationshipSatisfaction'] = min_max_scale(df['RelationshipSatisfaction'])
        df['Collaboration_Score'] = (df['N_JobInvolvement'] * 0.5 + df['N_RelationshipSatisfaction'] * 0.5)
        df.drop(columns=['N_JobInvolvement', 'N_RelationshipSatisfaction'], inplace=True)
        print(f"'Collaboration_Score' calculated for {df_name} (weak proxy).")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Skipping Collaboration_Score for {df_name}: required columns not found.")
````

**Reasoning**:
Following the feature engineering plan, I will now calculate the 'Sentiment_Score' for the `attendance` and `communication` DataFrames by directly using `satisfaction_level`. For the `tasks` and `feedback` DataFrames, I will combine `JobSatisfaction`, `EnvironmentSatisfaction`, and `RelationshipSatisfaction` to create the 'Sentiment_Score'.



````python
print("\n--- Calculating Sentiment Score ---")

# For attendance and communication DataFrames: using satisfaction_level directly
print("Calculating Sentiment_Score for attendance and communication (using satisfaction_level)...")
for df_name, df in [('attendance', attendance), ('communication', communication)]:
    if 'satisfaction_level' in df.columns:
        df['Sentiment_Score'] = df['satisfaction_level']
        print(f"'Sentiment_Score' calculated for {df_name}.")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Skipping Sentiment_Score for {df_name}: 'satisfaction_level' not found.")

# For tasks and feedback DataFrames: combining JobSatisfaction, EnvironmentSatisfaction, and RelationshipSatisfaction
print("\nCalculating Sentiment_Score for tasks and feedback (combining multiple satisfaction metrics)...")
for df_name, df in [('tasks', tasks), ('feedback', feedback)]:
    required_cols_sentiment = ['JobSatisfaction', 'EnvironmentSatisfaction', 'RelationshipSatisfaction']
    if all(col in df.columns for col in required_cols_sentiment):
        df['N_JobSatisfaction'] = min_max_scale(df['JobSatisfaction'])
        df['N_EnvironmentSatisfaction'] = min_max_scale(df['EnvironmentSatisfaction'])
        df['N_RelationshipSatisfaction'] = min_max_scale(df['RelationshipSatisfaction'])
        df['Sentiment_Score'] = (
            df['N_JobSatisfaction'] * 0.33 +
            df['N_EnvironmentSatisfaction'] * 0.33 +
            df['N_RelationshipSatisfaction'] * 0.34
        )
        df.drop(columns=['N_JobSatisfaction', 'N_EnvironmentSatisfaction', 'N_RelationshipSatisfaction'], inplace=True)
        print(f"'Sentiment_Score' calculated for {df_name}.")
        print(f"{df_name}.head():")
        display(df.head())
    else:
        print(f"Skipping Sentiment_Score for {df_name}: required columns {required_cols_sentiment} not found.")
````

## Prepare Data for Attrition Prediction

### Subtask:
Select features and the target variable from the `tasks` DataFrame, and convert the 'Attrition' target to a numerical format.


**Reasoning**:
To prepare the data for attrition prediction, I will select the specified features from the `tasks` DataFrame to create the feature set `X`, and extract the 'Attrition' column as the target variable `y`. Then, I will convert `y` into a numerical format, mapping 'No' to 0 and 'Yes' to 1, and display the head of both `X` and `y` to verify the transformation.



````python
print("Preparing features (X) and target (y) from the tasks DataFrame...")

# 1. Define a list of features (X_cols)
X_cols = [
    'Age',
    'DailyRate',
    'DistanceFromHome',
    'HourlyRate',
    'MonthlyIncome',
    'NumCompaniesWorked',
    'PercentSalaryHike',
    'TotalWorkingYears',
    'YearsAtCompany',
    'YearsInCurrentRole',
    'YearsSinceLastPromotion',
    'YearsWithCurrManager',
    'Productivity_Score',
    'Communication_Score',
    'Collaboration_Score',
    'Sentiment_Score'
]

# 2. Create the feature DataFrame X
X = tasks[X_cols].copy()
print(f"Feature DataFrame X created with {len(X_cols)} columns.")

# 3. Define the target variable y
y = tasks['Attrition'].copy()
print("Target variable y created from 'Attrition' column.")

# 4. Convert the 'Attrition' target variable y into a numerical format
# Map 'No' to 0 and 'Yes' to 1
y = y.map({'No': 0, 'Yes': 1})
print("Target variable 'Attrition' converted to numerical (0 for No, 1 for Yes).")

# 5. Display the first few rows of X and y to verify
print("\nX head:")
display(X.head())

print("\ny head:")
display(y.head())

print("Data preparation for attrition prediction completed.")
````

## Split Data into Training and Testing Sets

### Subtask:
Divide the prepared dataset into training and testing sets to evaluate the model's performance on unseen data. A standard split of 70% for training and 30% for testing will be used.


````python
from sklearn.model_selection import train_test_split

print("Splitting data into training and testing sets...")

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("Data split complete.")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

print("\nFirst 5 rows of X_train:")
display(X_train.head())
print("\nFirst 5 rows of y_train:")
display(y_train.head())
````

## Train RandomForestClassifier Model

### Subtask:
Initialize and train a `RandomForestClassifier` model using the training features (`X_train`) and target variable (`y_train`).


**Reasoning**:
To train a RandomForestClassifier model, I will import the necessary class, instantiate it with a specified random state for reproducibility, and then fit it to the provided training data `X_train` and `y_train`.



````python
from sklearn.ensemble import RandomForestClassifier

print("Initializing RandomForestClassifier model...")
# Initialize RandomForestClassifier with a fixed random_state for reproducibility
model = RandomForestClassifier(random_state=42)
print("Model initialized successfully.")

print("Training RandomForestClassifier model...")
# Fit the model to the training data
model.fit(X_train, y_train)
print("Model training completed.")
````

````python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

print("Making predictions on the test set...")
y_pred = model.predict(X_test)
print("Predictions made successfully.")

print("\n--- Model Performance Evaluation ---")

# 1. Calculate and print Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# 2. Generate and print Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# 3. Compute and display Confusion Matrix
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)

fig, ax = plt.subplots(figsize=(8, 6))
ConfusionMatrixDisplay.from_predictions(y_test, y_pred, cmap='Blues', ax=ax)
ax.set_title('Confusion Matrix for Attrition Prediction')
plt.tight_layout()
plt.show()

print("Model performance evaluation complete.")
````
